For the Variable *Gender Male/Female*, we will represent this variable using the numpy.random.choice distribution:
https://numpy.org/doc/stable/reference/random/generated/numpy.random.Generator.choice.html    

```python
import numpy as np
import matplotlib.pyplot as plt
#It's important to set a number where the algorithm of random values starts in order to get the same serie of numbers whenever we run the code
np.random.seed(1)

#g represents the total of trials which will be 100 and p will be the probability of success
x = np.random.choice(['Male', 'Female'], 100, p=[0.25, 0.75])

# Plot histogram with separate bins
plt.hist(x, bins=2, alpha=0.6, edgecolor='black')  # Add edgecolor='black' for black outlines
plt.ylabel('Frequency')
plt.title('Gender')
plt.show()
```

>Next we will represent the variable *'Age'* which will range between the ages from 18 to 24, for that we decided to plot the values to have a better view:

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('smh.csv')

# Extract values from 'Age' column
age = data['Age']

value_counts = age.value_counts()
#print(value_counts)

#plot values
plt.figure(figsize=(8, 6))
plt.hist(age, bins=20, color='skyblue', edgecolor='black')
```

>The highest values are concentrated around 18 and 24, decreasing towards the middle. To represent this distribution, we'll use random triangular distributions and we will combine both using concatenate distribution:

>https://modelassist.epixanalytics.com/display/EA/Triangular+distribution

>https://stackoverflow.com/questions/815969/python-simpy-how-to-generate-a-value-from-a-triangular-probability-distributio 

>https://numpy.org/doc/stable/reference/generated/numpy.concatenate.html

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

#we will create two different datasets with 50 values, the right dataset will decrease gradually and the left will increase by setting the mode to 18(L) and 25(R)
age_left = np.random.triangular(left=18, mode=18, right=21, size=50)
age_right = np.random.triangular(left=21, mode=25, right=25, size=50)


# now that we have the two distributions we will combine them using numpy.concatenate
combined_data = np.concatenate((age_left, age_right))
plt.hist(combined_data, bins=7, color='skyblue', edgecolor='black')
```

>Next, we will count how many students out of 100 suffer from Anxiety:

```python
import pandas as pd

data = pd.read_csv('smh.csv')

# Extract values from 'Do you have anxiety' column
gender = data['Do you have Anxiety?']

value_counts = gender.value_counts()
print(value_counts)
```

>Now we will create a random choice distribution with a probability 33% *Yes* and a 67% *No*

```python
import numpy as np
import matplotlib.pyplot as plt
#It's important set a number where the algorithm of random values starts in order to get the same serie of numbers
np.random.seed(1)

# Generate three values using binomial distribution
x = np.random.choice(['Yes', 'No'], 100, p=[0.33, 0.67])

# draw plot histogram with 2 separate bins 
plt.hist(x, bins=2, alpha=0.6, edgecolor='black')  
plt.ylabel('Number of Students')
plt.title('Students with Anxiety')
plt.show()
```

>Once we finished with the variables individually and created their random distributions, we can start establishing the relations between variables. First we will compare how much each Gender suffers from Anxiety:

>https://stackoverflow.com/questions/6871201/plot-two-histograms-on-single-chart

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

# Call gender data
genders = np.random.choice(['Male', 'Female'], 100, p=[0.25, 0.75])

# Call anxiety levels data
anxiety_levels = np.random.choice(['Yes', 'No'], 100, p=[0.33, 0.67])

# Plotting histogram for each gender's anxiety levels
plt.figure(figsize=(10, 6))

# bin display characteristics
bins = np.linspace(0, 1, 3)

# Plot for Male
plt.hist(anxiety_levels[genders == 'Male'], bins, alpha=1, edgecolor='black', label='Male')

# Plot for Female
plt.hist(anxiety_levels[genders == 'Female'], bins, alpha=0.6, edgecolor='black', label='Female')

# Adding labels and title
plt.xlabel('Anxiety')
plt.ylabel('Frequency')
plt.title('Anxiety Levels by Gender')
plt.legend()
plt.show()
```

>The results indicate the percentage of males affected by anxiety is half of those unaffected, whereas for females, the proportion of those experiencing anxiety is three-quarters of those not affected."

Next, we will compare how many students from each gender sought for a specialist treatment:

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(1)

# Call gender data
genders = np.random.choice(['Male', 'Female'], 100, p=[0.25, 0.75])

# Call anxiety levels data
help = np.random.choice(['Yes', 'No'], 100, p=[0.06, 0.94])

# Plotting histogram for each gender's anxiety levels
plt.figure(figsize=(10, 6))

# bin display characteristics
bins = np.linspace(0, 1, 3)

# Plot for Male
plt.hist(help[genders == 'Male'], bins, alpha=1, edgecolor='black', label='Male')

# Plot for Female
plt.hist(help[genders == 'Female'], bins, alpha=0.6, edgecolor='black', label='Female')

# Adding labels and title
plt.ylabel('Frequency')
plt.title('Specialist Treatment')
plt.legend()
plt.show()
```

>As we can see from the plot, even though male students suffer less from anxiety, they tend to search for a specialist treatment more often than female students.
